## Nhập thư viện và đọc dữ liệu

```python
#Import lib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_M1 = pd.read_excel('datasets/M1.xlsx', header = 3)
data_M2 = pd.read_excel('datasets/M2.xlsx', header = 3)
data_M3 = pd.read_excel('datasets/M3.xlsx', header = 3)
data_M4 = pd.read_excel('datasets/M4.xlsx', header = 3)
data_M5 = pd.read_excel('datasets/M5.xlsx', header = 3)
data_M6 = pd.read_excel('datasets/M6.xlsx', header = 3)
```

## Làm sạch dữ liệu

```python
data = [data_M1, data_M2, data_M3, data_M4, data_M5, data_M6]
features = ['STT','MSSV', 'Điểm']
def convert_to_int(data, features):
    for feature in features:
        data[feature] = data[feature].astype('int64')
for test in data:
    test.drop(test.index[0],inplace=True)
    convert_to_int(test, features)
    test.drop(['STT','NH','HK'],axis = 1, inplace = True)
```
